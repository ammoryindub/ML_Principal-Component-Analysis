## Machine Learning - Principal Component Analysis.

PCA is a data transformation technique that identifies the most important features explaining the variance in the data. It discards less significant components, focusing on the ones that contribute the most to the data's variability. 

### Import Libraries

```python
# starting with importing libariraies 

import numpy as np
import pandas as pd
```

```python
# importing data visualization libraries 

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

### Get the data 

I will create artificial data with sklearn capabilities.

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print (cancer['DESCR'])
```

after we got the data we need to set it up into a recognizable dataframe 

```python
df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
```

```python
df.head()
```

### PCA Visualization

To visualize high-dimensional data in a 2D scatter plot, we can use PCA to find the first two principal components. 

**N: Note that scaling the data to have unit variance in each feature is necessary before applying PCA.**

```python
from sklearn.preprocessing import StandardScaler
```

we need scale our data, so that each feature has a single unit variance before we actually use PCA

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

To perform dimensionality reduction using PCA, we instantiate a PCA object, find the principal components with the fit method, and then apply the rotation to reduce dimensions by calling transform() method.

### PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

30 dimensions in our scaled_data reduced to just 2 in x_pca. we plot these two dimensions now

```python
plt.figure (figsize =(12,8))
plt.scatter (x_pca[:,0], x_pca[:,1])

plt.xlabel('1st Principal Component')
plt.ylabel('2nd Principal Component')
```

```python
# now we try to add more arguments to make this easier to understand 

plt.figure (figsize =(12,8))
plt.scatter (x_pca[:,0], x_pca[:,1],c=cancer['target'],cmap='plasma')

plt.xlabel('1st Principal Component')
plt.ylabel('2nd Principal Component')
```

by using two components, we can separate these two classes

### Interpreting the components 

The components correspond to combinations of the original features, the components themselves are stored as an attribute of the fitted PCA object

```python
pca.components_
```

In this matrix array, each row represents a principal component, and each column relates back to the original features.

we can visualize this relationship with a heatmap

```python
df_comp = pd.DataFrame(pca.components_, columns= cancer['feature_names'])
```

```python
# heatmap

plt.figure (figsize =(12,8))
sns.heatmap (df_comp, cmap='plasma')
```

here have a heat map that shows the relationship between
the correlation of the various features and the principal component themselves.
